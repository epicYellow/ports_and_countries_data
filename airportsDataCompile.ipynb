```python
import pandas as pd
import json
```

```python
portsJson = 'ports.json'
ports = []
```

```python
with open(portsJson, 'r') as file:
    data = json.load(file)
    for key in data.keys():
        data_from_code = data[key]
        print({**data_from_code, "portCode": key})
        ports.append({**data_from_code, "portCode": key})
```

```python
for port in ports:
    for key in port.keys():
        print(key)
```

```python
df = pd.DataFrame(ports, columns=['name',
'city',
'country',
'coordinates',
'province',
'portCode'])

df
```

```python
df.to_csv('ports.csv', index=False)
```
